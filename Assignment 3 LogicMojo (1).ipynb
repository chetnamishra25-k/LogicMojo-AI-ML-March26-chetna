Q7. Feature Engineering for Model Improvement

Context: Raw data is rarely useful—features must be created.

Dataset: Titanic

Tasks:

1. Create:
family_size = sibsp + parch + 1
2. Create:
is_alone (binary feature)
3. Compute survival rate by is_alone
4. Explain:
Why engineered features improve ML models

```python
ds = sns.load_dataset("titanic")

def is_alone(x):
    if x == True:
        return 1
    else:
        return 0

#Create: family_size = sibsp + parch + 1
ds['family_size']=ds['sibsp']+ds['parch']+1

#Create: is_alone (binary feature)
# ds['is_alone'] = ds['alone'].map({True: 1, False: 0})
# ds['is_alone'] = ds['alone'].map(is_alone)
# ds['is_alone'] = ds['alone'].map(lambda x: 1 if x else 0)
ds['is_alone'] = ds['alone'].apply(is_alone)

#3. survival rate by is_alone
totalSurvived=ds.loc[ds['survived']==1].shape[0]
aloneSurvived= ds.loc[(ds['is_alone'] == 1) & (ds['survived'] == 1)].shape[0]
print("rate of survival for alone travellers :",aloneSurvived/totalSurvived)

#4. Engineered feature creates new meaning and hence 
#boost perfromance. Derives hidden meaning from raw data.
```

Q8. Group-Based Insights (Customer Segmentation Logic)

Context: ML models often depend on aggregated patterns.

Dataset: Tips

Tasks:

1. Compute:
Average tip by day
Total bill by gender
2. Create new feature:
tip_percentage
3. Identify:
Which day has highest tipping behavior
4. Explain:

```python
df = sns.load_dataset("tips")

print(df.head())

#1. Compute: Average tip by day
meanTip =df.groupby('day')['tip'].mean()
print(meanTip)

# Total bill by gender
billByGender =df.groupby('sex')['total_bill'].sum()
print(billByGender)

#2. Create new feature: tip_percentage
df['tip_percentage']=df['tip']/df['total_bill']*100
# print(df['tip_percentage'])

#3. Identify: Which day has highest tipping behavior
highest_day = meanTip.idxmax() #👉 idxmax() returns the index (label) of the maximum value in a Series or DataFrame column.
highest_value = meanTip.max()

print("Highest tipping day:", highest_day)
print("Average tip:", highest_value)

#4 segmentation model
# These help in grouping data into meaningful cluster or categories.
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
ds = sns.load_dataset("titanic")
# print(ds)

ds['age'].fillna(ds['age'].median(), inplace=True)


def age_group(age):
    if(age<18):
        return "Child"
    elif(18<=age<55):
        return "Adult"
    else:
        return "Senior"

#Create age_group: Child / Adult / Senior
ds['age_group']=ds['age'].apply(age_group)
# print(ds)

#Compute survival rate per group

survival_rate = ds.groupby("age_group")["survived"].mean()

print(survival_rate)

#4 Child has highest survival likelihood
```
