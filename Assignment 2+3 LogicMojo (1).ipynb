## 1. Customer Segmentation Engine

You are given a list of customer dictionaries:

```python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
```

Task:

1. Write a function segment_customer(customer):
2. If spend divided by income is greater than 0.6 return "High Value"
3. If spend divided by income is between 0.3 and 0.6 return "Medium Value"
4. Otherwise return "Low Value"
5. Create a final dictionary in the format {customer_id: segment}

````python
def segment_customer(customer):
    id,age,income,spend=customer.values()
    ratio = int(spend)/int(income)
    if(ratio>0.6):
        return "High Value"
    elif (0.3<ratio<0.6):
        return "Medium Value"
    else:
        return "low Value"

customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]

results =[]
for customer in customers:
    id,age,income,spend =customer.values()
    segment= segment_customer(customer)
    customer_seg = { "customer_id":id ,"segment": segment}
    results.append(dict(customer_seg))
print(results)
    
````

4. Use lambda with filter or map

## 6. Sequence Pattern Detection

You are given:

```python
data = [10, 20, 30, 25, 40, 50]
```

Task:

1. Write a function detect_increasing_sequence(data)
2. Find the longest continuous increasing sequence
3. Return that sequence as a list

Example output:

```python
[10, 20, 30]
```


````python
data = [10, 20, 30, 25, 40, 50]
# data = [10, 20, 30]

def detect_increasing_sequence(data):
    longest_seq = []
    current_seq = [data[0]]
    for i in range(1, len(data)):
        if data[i] > data[i - 1]:
            # increasing → add to current sequence
            current_seq.append(data[i])
        else:
            # sequence broke → check if current is longest
            if len(current_seq) > len(longest_seq):
                longest_seq = current_seq
            # start new sequence
            current_seq = [data[i]]

    return longest_seq

print(detect_increasing_sequence(data))
````

## 7. Dictionary-Based Recommendation System

You are given:

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
```

Task:

1. Write a function recommend(user)
2. Recommend items based on common items between users
3. Do not recommend items already purchased by the user
4. Use sets and dictionary iteration

---

````python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}

# def recommend(user):

s = set()
    
for key, value in user_purchases.items():
    s.update(value)

results =[]
    
def recommend(users):
    for key, value in users.items():
        suggestion= s-set(value)
        user = { key : value, "suggestion": suggestion}
        results.append(user)
    return results


print(recommend(user_purchases))
````
